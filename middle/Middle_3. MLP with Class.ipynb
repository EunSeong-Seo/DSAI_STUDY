```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

raw_data = load_iris()

X_data = raw_data.data
y_data = raw_data.target

X_train, X_test, y_train, y_test = train_test_split(X_data,y_data,test_size=0.3,random_state =7 )

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#데이터 정규화 및 변환
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test - scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from tensorflow.keras import optimizers

class MLP:
  def __init__(self ):
    self.model = Sequential()
  def make_layers(self,h, input_shape ,act):
    self.model.add(Dense(h,input_shape = (input_shape,)))
    self.model.add(Activation(act))
    self.model.add(Dense(h))
    self.model.add(Activation(act))
    self.model.add(Dense(h))
    self.model.add(Activation(act))
    self.model.add(Dense(1))
    self.model.add(Activation("sigmoid"))
  def select_opt(self,lr, loss, metrics):
    sgd = optimizers.SGD(lr = lr)
    self.model.compile(optimizer = sgd, loss = loss, metrics = [metrics])
  def train(self,X,y,bat_size,ep):
    self.model.fit(X,y,batch_size = bat_size, epochs = ep, verbose=1)
  def result(self, X_t,y_t):
    results = self.model.evaluate(X_t,y_t)
    print(self.model.metrics_names)
    print(results)

    print('loss : ',results[0])
    print('accuracy : ',results[1])
```

```python
#모델 생성
two_layers_MLP = MLP()
two_layers_MLP.make_layers(10,4,'sigmoid')
two_layers_MLP.select_opt(0.01,'binary_crossentropy','acc')
```

```python
#모델 학습
two_layers_MLP.train(X_train,y_train,50,100)
```

```python
#모델 결과
two_layers_MLP.result(X_test,y_test)
```
